```python
## The MSE of the general model

mse_train = 420092468.66268444
```

```python
## Find out the MSE of every modal built on data from each neighbourhood

neighborhood = df['Neighborhood'].unique()
per_dif = []
for i in neighborhood:

  df1 = df[df['Neighborhood']==i]
  df1_sale = df1['SalePrice']
  df1 = df1.drop(columns=['SalePrice'])
  
  df1 = preprocessing(df1)

  ridge = Ridge(alpha=40)
  ridge_test = ridge.fit(df1, df1_sale)
  y_pred_test = ridge_test.predict(df1)

  mse_df1 = mean_squared_error(df1_sale, y_pred_test)
  perc_diff = abs((mse_df1 - mse_train)/mse_df1 )* 100
  per_dif.append(perc_diff)
  print('Neighbourhood: ',i, 'with mse of',mse_df1, 'different from general model by:',str(perc_diff)+"%")
```

```python
neigh_data = pd.concat([pd.DataFrame(neighborhood, columns=['Neighbourhood']),pd.DataFrame(per_dif, columns=['MSE diff with baseline'])],axis=1)
neigh_data.sort_values(by=['MSE diff with baseline'],ascending=True).head()
```

From the result, OldTown, CollegeCreek and Somerst are choosen due to their low MSE diff with train data and sufficient number of data available (>100).

## Next, start investigating the coeffiecient of each model on the 3 neighbourhood to find out how the features associate with the property sale price

#### First, generate 3 different models which each model is built base on the corresponding neighbourhood data. The features which having the highest coefficient in our general training model are used in our analysis:
'Lot Area', 'Gr Liv Area','Bsmt Exposure','Total Bsmt SF', 'Bedroom AbvGr', 'Bsmt Full Bath', 'Bsmt Half Bath' ,'Full Bath' ,'Half Bath','Kitchen AbvGr','Year Built'

```python
## Generate 3 models on 3 neighbourhoods

Neigh = ['OldTown','CollgCr','Somerst']
feature_neigh = ['Lot Area', 'Gr Liv Area','Bsmt Exposure','Total Bsmt SF', 'Bedroom AbvGr', 'Bsmt Full Bath', 'Bsmt Half Bath' ,'Full Bath' ,'Half Bath',
                 'Kitchen AbvGr','Year Built']
raw_dic = {}
Neigh_dic = {}
columns_name = pickle.load(open('./columns_name.pkl', 'rb'))

for i in Neigh: 
  
  df_Coll = df[df['Neighborhood']==i]
  df_Coll_sale = df_Coll['SalePrice']
  df_Coll_test  = df_Coll.drop(columns=['SalePrice'])
  df_Coll_test = preprocessing(df_Coll_test)
  raw_dic[i] = df_Coll_test

  ridge_Coll = Ridge(alpha=40)
  ridge_Coll_test = ridge_Coll.fit(df_Coll_test , df_Coll_sale)

  model_Coll = pd.DataFrame(ridge_Coll.coef_, columns=['Coef'])
  model_Coll['feature'] = columns_name
  
  iteration = []
  for feature in feature_neigh:
      feat =  model_Coll.loc[model_Coll['feature']==feature]['Coef'].values[0]
      iteration.append(feat)

  data = pd.DataFrame(feature_neigh)
  data['Coef'] = iteration
    
  Neigh_dic[i] = data    
  print('Model Generated base on Neighborhood ',i, 'done!')
```

#### In order to calculate the investment ROI, our team have taken the average cost to build the rooms from scratch.
Src: https://homeguide.com/costs/

```python
## Define the average cost of remodelling each type of room
cost_bath = 7600
cost_half_bath = 5000
cost_bed = 12100
```

### A relative coefficient of the target feature is computed by considering the correlation with all other features This steps are necessary as the linear model is not fully eliminate the collinearity problem 

```python
def relative_feature(feature):

  oldTown_df = Neigh_dic['OldTown']
  oldtown_df1 = raw_dic['OldTown']
  old_town_corr = oldtown_df1[feature_neigh].corr()
  old_town_corr_f = old_town_corr[old_town_corr.index==feature]
  oldTown_df['Rel_Coef_oldtown'] = oldTown_df['Coef']* old_town_corr_f.T[feature].values

  CollgCr_df = Neigh_dic['CollgCr']
  CollgCr_df1 = raw_dic['CollgCr']
  CollgCr_corr = CollgCr_df1[feature_neigh].corr()
  CollgCr_corr_f = CollgCr_corr[CollgCr_corr.index==feature]
  CollgCr_df['Rel_Coef_CollgCr'] = CollgCr_df['Coef']* CollgCr_corr_f.T[feature].values

  Somerst_df = Neigh_dic['Somerst']
  Somerst_df1 = raw_dic['Somerst']
  Somerst_corr = Somerst_df1[feature_neigh].corr()
  Somerst_corr_f = Somerst_corr[Somerst_corr.index==feature]
  Somerst_df['Rel_Coef_Somerst'] = Somerst_df['Coef']* Somerst_corr_f.T[feature].values

  combine_data = oldTown_df.merge(CollgCr_df, on=0)
  combine_data = combine_data.merge(Somerst_df, on=0)
  combine_data.set_index(0,inplace=True)
  combine_data=combine_data.sort_values(by=['Rel_Coef_Somerst'],ascending=False)
  
  return combine_data
```

## 1st Feature: The effect of number of Bedrooms on salesprice and the ROI

```python
## Generate a relative data to bedrooms
bed_data = relative_feature('Bedroom AbvGr')
```

```python
fig, ax = plt.subplots(figsize=(18,8))
bed_data[['Rel_Coef_oldtown','Rel_Coef_CollgCr','Rel_Coef_Somerst']].plot(kind='bar', stacked=False,width=0.8,color=["indigo","blue", "cyan", ],ax=ax)
plt.legend(['Oldtown','College Creek','Sommert'],fontsize=15)
plt.grid()
plt.title('Sales Price increase by each Feature per unit \n Relative to Number of Bedrooms',fontsize=20)
plt.xlabel('Features',fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=12)
plt.ylabel('Change in Sales Price, $',fontsize=15)
```

The relative factors are all included to get the real effect of how increasing number of rooms affect the sales price

```python
num_bed = np.arange(1,10) #number of bedroom to be simulated

#including all other collinear features' coefficient 
bed_ot = bed_data['Rel_Coef_oldtown'].sum() 
bed_cl = bed_data['Rel_Coef_CollgCr'].sum() 
bed_sm = bed_data['Rel_Coef_Somerst'].sum() 

#generate salesprice estimation and ROI
#factor in standard scaler to our features here
y_bed_ot = bed_ot*num_bed
y_bed_cl = bed_cl*num_bed
y_bed_sm = bed_sm*num_bed

ROI_bed_ot = ((y_bed_ot-cost_bed*num_bed)*100/(cost_bed*num_bed)).mean()
ROI_bed_cl = ((y_bed_cl-cost_bed*num_bed)*100/(cost_bed*num_bed)).mean()
ROI_bed_sm = ((y_bed_sm-cost_bed*num_bed)*100/(cost_bed*num_bed)).mean()
```

```python
ROI_bed_sm
```

```python
fig, ax = plt.subplots(figsize=(18,10))
ax.plot(num_bed, y_bed_ot,marker='o',color="indigo",linewidth=2, markersize=5)
ax.plot(num_bed, y_bed_cl,marker='s',color="blue",linewidth=2, markersize=5)
ax.plot(num_bed, y_bed_sm,marker='v',color="cyan",linewidth=2, markersize=5)
ax.set_ylim([0, 200000])
ax.grid()
plt.title('Sales Price increase by number of Bedrooms & \nThe ROI of adding Bedrooms',fontsize=20)
ax.set_xlabel('Number of Bedrooms',fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_ylabel('Change in Sales Price, $',fontsize=15)

width=0.25
ax2 = ax.twinx()
ax2.bar(num_bed-width, ROI_bed_ot, color = "indigo",
        width = width, edgecolor = 'black',
        label='old town')
ax2.bar(num_bed , ROI_bed_cl, color = "blue",
        width = width, edgecolor = 'black',
        label='CollCr')
ax2.bar(num_bed + width, ROI_bed_sm, color = "cyan",
        width = width, edgecolor = 'black',
        label='Somerst')
plt.xticks(num_bed) 
ax2.set_ylim([0, 150])
ax2.legend(fontsize=15)
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.set_ylabel('ROI of adding bedrooms, %',fontsize=15)
```

Findings: The result shows that all neighbourhood property having positive ROI but Sommerset property seems to have higher ROI
