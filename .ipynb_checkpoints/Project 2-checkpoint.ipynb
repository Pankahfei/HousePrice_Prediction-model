```python
df_train = df_train.drop(columns='Id')
df_test = df_test.drop(columns='Id')
df_train.head()
```

```python
df_null = pd.DataFrame((100*df_train.isnull().sum()/len(df_train)),columns=['percentage']).sort_values(by='percentage',ascending=False)
df_null = df_null[df_null['percentage']>0]
df_null.shape
```

```python
df_null.head(20)
```

```python
f, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=df_null.index, y=[j for value in df_null.values for j in value])
plt.xticks(rotation='90')
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percentage of missing values', fontsize=15)
plt.title('Percentage of missing data by feature', fontsize=15)
plt.axhline(y = 3, color = 'r', linestyle = '-')
```

```python
df_null_test = pd.DataFrame((100*df_test.isnull().sum()/len(df_test)),columns=['percentage']).sort_values(by='percentage',ascending=False)
df_null_test = df_null_test[df_null_test['percentage']>0]
df_null_test.index
```

```python
f, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=df_null_test.index, y=[j for value in df_null_test.values for j in value])
plt.xticks(rotation='90')
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percentage of missing values', fontsize=15)
plt.title('Percentage of missing data by feature', fontsize=15)
plt.axhline(y = 3, color = 'r', linestyle = '-')
plt.show()
```

```python
y = df_train['SalePrice']

(mu,sigma) = norm.fit(y)
print(f'mu = {round(mu,2)}, sigma = {round(sigma,2)}, skew = {y.skew()}')
f, ax = plt.subplots(figsize=(10,5))
ax = sns.distplot(y, fit=norm)
plt.xlabel('SalePrice', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Distribution of Housing Sale Price', fontsize=15)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(df_train.corr(),cmap='Blues',annot=True)
```

```python
df_corr = df_train.corr()
top_corr_features = list(df_corr[(df_corr['SalePrice']>0.6)&(df_corr['SalePrice']!=1)].index)
top_corr_features.append('Year Built')
```

```python
top_corr_plot = pd.concat([y,df_train[top_corr_features]],axis=1)
sns.pairplot(data=top_corr_plot)
```

```python
linear_outliers = ['Total Bsmt SF','1st Flr SF','Gr Liv Area','Garage Area','Year Built' ]

fig, ax = plt.subplots(figsize=(12, 12))
plt.subplots_adjust(right=2)
plt.subplots_adjust(top=2)

for i, feature in enumerate(linear_outliers,1):

    plt.subplot(len(linear_outliers), 2, i)
    sns.scatterplot(x=feature, y='SalePrice', palette='Blues', data=top_corr_plot)     
    plt.xlabel('{}'.format(feature), size=15,labelpad=12.5)
    plt.ylabel('SalePrice', size=15, labelpad=12.5)
    
    for j in range(2):
        plt.tick_params(axis='x', labelsize=15)
        plt.tick_params(axis='y', labelsize=15)
        
plt.show()
```

```python
Categorical_outliers = ['Overall Qual','Garage Cars']

fig,ax = plt.subplots(figsize=(20,12))
plt.subplots_adjust(right=2)
plt.subplots_adjust(top=2)

for i, feature in enumerate(Categorical_outliers,1):
    
    plt.subplot(len(Categorical_outliers), 1, i)
    sns.boxplot(x=feature, y='SalePrice', palette='Blues', data=top_corr_plot)     
    plt.xlabel('{}'.format(feature), size=25,labelpad=12.5)
    plt.ylabel('SalePrice', size=25, labelpad=12.5)
    for j in range(2):
        plt.tick_params(axis='x', labelsize=25)
        plt.tick_params(axis='y', labelsize=25)

plt.show()
```

## Data Cleaning
- Decide how to impute null values.
- Decide how to handle outliers.
- Do you want to combine any features?
- Do you want to have interaction terms?
- Do you want to manually drop collinear features?

```python
df_train.drop(columns=list(df_null[df_null['percentage']>3].index),inplace=True)
null_to_drop = list(df_null[df_null['percentage']<=3].index)
for i, feature in enumerate(null_to_drop):
    df_train = df_train.drop(df_train[df_train[feature].isnull()].index)
```

```python
df_train.shape
```

```python
df_train.isnull().sum().sort_values(ascending=False)
```

```python
df_train.drop(df_train[(df_train['SalePrice']<300000)&(df_train['Total Bsmt SF']>4000)].index, inplace=True)
df_train.drop(df_train[(df_train['SalePrice']<300000)&(df_train['1st Flr SF']>4000)].index, inplace=True)
df_train.drop(df_train[(df_train['SalePrice']<300000)&(df_train['Gr Liv Area']>4000)].index, inplace=True)
```

```python
linear_outliers = ['Total Bsmt SF','1st Flr SF','Gr Liv Area','Garage Area','Year Built' ]

fig, ax = plt.subplots(figsize=(12, 12))
plt.subplots_adjust(right=2)
plt.subplots_adjust(top=2)

for i, feature in enumerate(linear_outliers,1):

    plt.subplot(len(linear_outliers), 2, i)
    sns.scatterplot(x=feature, y='SalePrice', palette='Blues', data=df_train)     
    plt.xlabel('{}'.format(feature), size=15,labelpad=12.5)
    plt.ylabel('SalePrice', size=15, labelpad=12.5)
    
    for j in range(2):
        plt.tick_params(axis='x', labelsize=15)
        plt.tick_params(axis='y', labelsize=15)
        
plt.show()
```

```python
df_train.drop(df_train[(df_train['SalePrice']>400000)&(df_train['Overall Qual']==6)].index, inplace=True)
df_train.drop(df_train[(df_train['SalePrice']>500000)&(df_train['Overall Qual']==8)].index, inplace=True)
df_train.drop(df_train[(df_train['SalePrice']>300000)&(df_train['Garage Cars']==1)].index, inplace=True)
df_train.drop(df_train[(df_train['SalePrice']>400000)&(df_train['Garage Cars']==4)].index, inplace=True)
```

```python
Categorical_outliers = ['Overall Qual','Garage Cars']

fig,ax = plt.subplots(figsize=(20,12))
plt.subplots_adjust(right=2)
plt.subplots_adjust(top=2)

for i, feature in enumerate(Categorical_outliers,1):
    
    plt.subplot(len(Categorical_outliers), 1, i)
    sns.boxplot(x=feature, y='SalePrice', palette='Blues', data=df_train)     
    plt.xlabel('{}'.format(feature), size=25,labelpad=12.5)
    plt.ylabel('SalePrice', size=25, labelpad=12.5)
    for j in range(2):
        plt.tick_params(axis='x', labelsize=25)
        plt.tick_params(axis='y', labelsize=25)

plt.show()
```
