## Pre-processing
- One-hot encode categorical variables.
- Train/test split your data.
- Scale your data.
- Consider using automated feature selection.


```python
X = df_train.drop(columns='SalePrice')
y = np.log1p(df_train['SalePrice'])
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
print(f'Before feature engineering, X training data shape:{X_train.shape},X testing data shape:{X_test.shape}')
```

```python
X_train.reset_index(drop=True,inplace=True)
X_test.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)
```

```python
numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
def VIF_calculator(data):
  numeric = []
  for i in data.columns:
     if data[i].dtype in numeric_dtypes:
        numeric.append(i)
  X_coll = data[numeric]  
  vif_data = pd.DataFrame()
  vif_data['feature'] = X_coll.columns
  vif_data['VIF'] = [variance_inflation_factor(X_coll.values, i) for i in range(len(X_coll.columns))]

  return vif_data[vif_data['VIF']>10]
```

```python
VIF_calculator(X_train)    
```

```python
X_train['Total_garagearea'] = X_train['Garage Cars'] + X_train['Garage Area']
X_train['Quality&Cond'] = (X_train['Overall Qual']+X_train['Overall Cond'])
X_train = X_train.drop(columns=['Year Remod/Add','TotRms AbvGrd','Garage Cars','Garage Area','Overall Qual','Overall Cond',
                                'BsmtFin SF 1','BsmtFin SF 2','Bsmt Unf SF','1st Flr SF','2nd Flr SF','Low Qual Fin SF'])
```

```python
VIF_calculator(X_train)   
```

```python
def label_encode(data): 
 encode_feature=[]
 lbl = LabelEncoder()
 for i in data.columns:
    if data[i].dtype == object:
        encode_feature.append(i)
 for n in encode_feature:      
    data[n] = lbl.fit_transform(list(data[n].values))
 return data
```

```python
X_train = label_encode(X_train)
sc = StandardScaler()
Z_train = pd.DataFrame(sc.fit_transform(X_train), columns=X_train.columns)
```

```python
X_test['Total_garagearea'] = X_test['Garage Cars'] + X_test['Garage Area']
X_test['Quality&Cond'] = (X_test['Overall Qual']+X_test['Overall Cond'])
X_test = X_test.drop(columns=['Year Remod/Add','TotRms AbvGrd','Garage Cars','Garage Area','Overall Qual','Overall Cond',
                                'BsmtFin SF 1','BsmtFin SF 2','Bsmt Unf SF','1st Flr SF','2nd Flr SF','Low Qual Fin SF'])
```

```python
X_test = label_encode(X_test)
Z_test = pd.DataFrame(sc.transform(X_test),columns=X_test.columns)
```

```python
print(f'After feature engineering, X training data shape:{Z_train.shape},X testing data shape:{Z_test.shape} while y training data shape:{y_train.shape}, y testing data shape:{y_test.shape}')
```

```python
X_statmodel = sm.add_constant(Z_train)
model = sm.OLS(y_train,X_statmodel)
result = model.fit()
print(result.summary())
```

```python
result.params.sort_values(ascending=False)
```

## Modeling
- **Establish your baseline score.**
- Fit linear regression. Look at your coefficients. Are any of them wildly overblown?
- Fit lasso/ridge/elastic net with default parameters.
- Go back and remove features that might be causing issues in your models.
- Tune hyperparameters.
- **Identify a production model.** (This does not have to be your best performing Kaggle model, but rather the model that best answers your problem statement.)
- Refine and interpret your production model.

```python
lr = LinearRegression()
lr.fit(Z_train, y_train)
y_train_pred_lr = lr.predict(Z_train)
y_test_pred_lr = lr.predict(Z_test)
```

```python
mse_train = mean_squared_error(y_train, y_train_pred_lr)
mse_test = mean_squared_error(y_test, y_test_pred_lr)

perc_diff = (mse_train - mse_test)/mse_train * 100
print(mse_train, mse_test, str(perc_diff)+"%")
```

```python
l_alphas=np.logspace(-3,10,100)

lasso_cv = LassoCV(alphas=l_alphas, cv=3, max_iter=50000)
lasso_model = lasso_cv.fit(Z_train, y_train)
y_train_pred_la = lasso_model.predict(Z_train)
y_test_pred_la = lasso_model.predict(Z_test)

print(lasso_cv.alpha_)
mse_train = mean_squared_error(y_train, y_train_pred_la)
mse_test = mean_squared_error(y_test, y_test_pred_la)

perc_diff = (mse_train - mse_test)/mse_train * 100
print(mse_train, mse_test, str(perc_diff)+"%")
```
